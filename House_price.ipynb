# **Bivariate Analysis**



 Price vs Total Square Feet (scatter + regression)


```python
sns.set_style("whitegrid")

plt.figure(figsize=(8,5))
sns.regplot(data=df, x='total_sqft_clean', y='price',
            scatter_kws={'alpha':0.3, 'color':'blue'},
            line_kws={'color':'red', 'linewidth':2})
plt.xlabel('Total Square Feet')
plt.ylabel('Price (Lakhs)')
plt.xlim(0, df['total_sqft_clean'].quantile(0.99))  # optional: zoom to 99th percentile
plt.ylim(0, df['price'].quantile(0.99))
plt.show()
```

1. The Scatter plot shows a clear positive upward linear regression
2. Most properties fall in the 800–2500 sqft range with prices up to 150 lakhs.
3. However, there is considerable spread, indicating that size alone does not determine price – location, bathrooms, and other features also play a role. Several outliers exist above the regression line, representing premium properties."

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df, x='bath', y='price', palette='Set2')
```

1. Price generally increases with more bathrooms (up to a point)

# **MULTIVARIATE ANALYSIS**

```python
multivariate_df = df[['price', 'total_sqft_clean', 'bath', 'rooms', 'balcony',
                      'area_type', 'availability', 'location']].copy()

multivariate_df.dropna(inplace=True)

top_locs = multivariate_df['location'].value_counts().head(10).index
multivariate_df['location_grouped'] = multivariate_df['location'].apply(lambda x: x if x in top_locs else 'Other')
```

```python
sns.pairplot(multivariate_df[['price', 'total_sqft_clean', 'bath', 'rooms']],
             diag_kind='hist', height=2.5)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=multivariate_df, x='total_sqft_clean', y='price',
                hue='bath', palette='viridis', alpha=0.6)
```

```python
g = sns.FacetGrid(multivariate_df, col='area_type', col_wrap=2, height=4, sharex=False)
g.map(sns.scatterplot, 'total_sqft_clean', 'price', alpha=0.5)
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=multivariate_df, x='location_grouped', y='price',
            order=multivariate_df.groupby('location_grouped')['price'].median().sort_values(ascending=False).index)
plt.xticks(rotation=45)
```

# **Summary of Multivariate Anaysis**
1. Multivariate analysis confirms that total square footage, bathroom count, area type, and location are the most important determinants of house price in Bengaluru. The relationships are positive, moderate to strong, and together explain about 70‑80% of the price variation. Balcony and extreme room counts are not useful predictors.
2. The same sqft, houses with more bathrooms are generally more expensive.
3. Recommendation: Use only bath (or only rooms) in predictive models.
4.  Plot Area properties tend to have higher prices for the same sqft compared to Super built‑up Area.
5. Top locations (Whitefield, Electronic City, Yelahanka) had significantly higher median prices.
6. Balcony has weak influence


# **Over all Summary of Analysis**
1. Size matters most – each additional 100 sqft adds roughly ₹8,000–12,000 to price.

2. Each extra bathroom adds ₹10–20 lakhs, but usually comes with larger size.

3. Plot Area commands a premium over built‑up area (land value).

4. Location is critical – Whitefield, Electronic City, Yelahanka are among the most expensive.

5. Balcony does not significantly affect price.

6. Bathrooms and rooms are highly correlated – don’t use both together in analysis.
